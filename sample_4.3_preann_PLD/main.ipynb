```python
fig = plt.figure(figsize=(12, 5))
gs = GridSpec(ncols=2, nrows=1, figure=fig)

ax_1 = fig.add_subplot(gs[0])
ax_2 = fig.add_subplot(gs[1])

y_major = ticker.LogLocator(numticks = 20)
y_minor = ticker.LogLocator(subs = np.arange(1.0, 10.0) * 0.1, numticks = 20)

d.draw_multiple_lines('C3-100', ax_1, measurs = [i for i in range(15, 80)], linewidth = 0.5, color = '#a3a3a3')
d.draw_single_line('C3-100', ax_1, measur=32, color = 'k', linewidth = 2)
ax_1.set(yscale = 'log', ylim=[10**(-13), 10**(-4)], xlim=[-1.25, 4.25])
ax_1.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax_1.xaxis.set_minor_locator(ticker.MultipleLocator(0.25))
ax_1.yaxis.set_major_locator(y_major)
ax_1.yaxis.set_minor_locator(y_minor)
ax_1.tick_params(labelsize=15)
ax_1.set_xlabel(', ', size = 17, fontfamily = 'serif')
ax_1.set_ylabel(', ', size = 17, fontfamily = 'serif')
ax_1.text(2, 2 * 10 ** -5, r'$I_{c.c.} = 10^{-5} \;A$', size = 13)

ax_1.text(-2.5, 1e-4, 'a)', size = 17)

d.draw_multiple_lines('G4-100', ax_2, measurs = [4,5,7,8,9,10,11,19], linewidth = 0.5, color = '#a3a3a3')
ax_2.set(yscale = 'log', ylim=[10**(-13), 10**(-4)], xlim=[-1.25, 2.25])
ax_2.xaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax_2.xaxis.set_minor_locator(ticker.MultipleLocator(0.1))
ax_2.yaxis.set_major_locator(y_major)
ax_2.yaxis.set_minor_locator(y_minor)
ax_2.tick_params(labelsize=15)
ax_2.set_xlabel(', ', size = 17, fontfamily = 'serif')
ax_2.set_ylabel(', ', size = 17, fontfamily = 'serif')
ax_2.text(1, 2 * 10 ** -6, r'$I_{c.c.} = 10^{-6} \;A$', size = 13)

ax_2.text(-2, 1e-4, ')', size = 17)

gs.update(wspace=0.3)
plt.savefig('preann_PLD_typical.png', dpi = 300, bbox_inches = 'tight')
```
