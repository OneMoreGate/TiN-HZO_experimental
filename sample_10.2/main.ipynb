```python
fig, ax = plt.subplots(ncols=3, figsize = [15,5])
A17_100_voltages = d_1.get_ts_on_off_voltage('A17-100', 'p', 0.3)
A17_100_currents = d_1.get_on_off_current('A17-100', 0.3, measurs=list(range(40,171)))
d_1.draw_multiple_lines('A17-100', ax[0], measurs=list(range(40,171)), linewidth = 0.3, color = 'gray')
ax[0].set_yscale('log')
ax[0].set_xlim([-1.1,1.1])
ax[1].hist(A17_100_voltages[0][40:], bins = 9)
ax[1].hist(A17_100_voltages[1][40:], bins = 4)


I_on = []
I_off = []
for i in range(len(A17_100_currents['off'])):
    if A17_100_currents['on'][i] != 0:
        I_on.append(A17_100_currents['off'][i])
        I_off.append(A17_100_currents['on'][i])
bins_1 = np.logspace(np.log10(min(np.abs(I_on))), np.log10(max(np.abs(I_on))), 10)
bins_2 = np.logspace(np.log10(min(np.abs(A17_100_currents['on']))), np.log10(max(np.abs(A17_100_currents['on']))), 20)

ax[2].hist(list(A17_100_currents['off']), bins = bins_1)
ax[2].hist(list(A17_100_currents['on']), bins = bins_2)
ax[2].set_xscale('log')
ax[2].relim()
```
