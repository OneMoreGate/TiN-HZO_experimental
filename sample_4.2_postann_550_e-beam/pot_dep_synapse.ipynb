```python
df = pd.read_csv('sample_4.2_7/I16-100/_019.txt')

fig, ax = plt.subplots(ncols=1, nrows=2, figsize = [15, 7])
ax[0].plot(df['Time'], df['Voltage'])
#ax[0].scatter(df['Time'], df['Voltage'])
#ax[0].set_xlim([0, 0.5e-6])
ax[1].plot(df['Time'], df['Current'])
#ax[1].scatter(df['Time'], df['Current'])
#ax[1].set_xlim([0, 0.5e-6])
#ax[0].set_title('sample_4.2_7/I16-100/_019')
#plt.savefig('sample_4.2_7_I16-100__019.png', dpi = 300, bbox_inches = 'tight')
```

```python
n_periods = 30
T_peroid = df['Time'][len(df['Time']) - 1] / n_periods
dots_per_period = int(len(df['Time']) / n_periods)
T_peroid, dots_per_period
```

```python
periods_passed = 0
average_resistance = []
time = []
for i in range(n_periods):
    dots = list(range(periods_passed * dots_per_period + 25, periods_passed * dots_per_period + 35))
    df_current_slice = np.array(df.iloc[dots ,2])
    df_voltage_slice = np.array(df.iloc[dots ,1])
    time.append(df.iloc[periods_passed * dots_per_period + 25 ,0])
    average_resistance.append(np.abs(np.mean(df_voltage_slice / df_current_slice)))
    periods_passed += 1
```

```python
fig, ax = plt.subplots(ncols=1, nrows=3, figsize = [10, 10])

plt.rcParams.update({'font.family':'serif'})

ax[0].plot(np.array(df['Time']) * 1e6, df['Voltage'])
ax[0].tick_params(labelsize = 13)
ax[0].xaxis.set_major_locator(ticker.MultipleLocator(1))
ax[0].xaxis.set_minor_locator(ticker.MultipleLocator(0.5))
ax[0].set_ylabel('Voltage, V', size = 15)

ax[1].plot(np.array(df['Time']) * 1e6, np.array(df['Current']) * 1e3)
ax[1].tick_params(labelsize = 13)
ax[1].xaxis.set_major_locator(ticker.MultipleLocator(1))
ax[1].xaxis.set_minor_locator(ticker.MultipleLocator(0.5))
ax[1].yaxis.set_major_locator(ticker.MultipleLocator(5))
ax[1].yaxis.set_minor_locator(ticker.MultipleLocator(2.5))
ax[1].set_ylabel('Current, mA', size = 15)

ax[2].scatter(np.array(time) * 1e6 , average_resistance)
ax[2].tick_params(labelsize = 13)
ax[2].xaxis.set_major_locator(ticker.MultipleLocator(1))
ax[2].xaxis.set_minor_locator(ticker.MultipleLocator(0.5))
ax[2].yaxis.set_major_locator(ticker.MultipleLocator(200))
ax[2].yaxis.set_minor_locator(ticker.MultipleLocator(100))
ax[2].set_ylabel('Resistance, Ohm', size = 15)
ax[2].set_xlabel(r'Time, $\mu s$', size = 15)

#plt.savefig('depression.png', dpi = 300, bbox_inches = 'tight')
```
